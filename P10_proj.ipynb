```python
import pandas as pd
df1 = pd.read_csv('countries.csv')
```

```python
df1.head()
```

```python
import sqlite3
conn = sqlite3.connect('countries.db')
```

```python
df1['Literacy (%)'] = [float(str(i).replace(',','.')) for i in list(df1['Literacy (%)'])]
df1['Agriculture'] = [float(str(i).replace(',','.')) for i in list(df1['Agriculture'])]
df1['Infant mortality (per 1000 births)'] = [float(str(i).replace(',','.')) for i in list(df1['Infant mortality (per 1000 births)'])]
df1.head()
```

```python
df1.to_sql('country', conn, if_exists='replace', index=False)
```

```python
def select_asian_countries(conn, region, population):# All Asian countries with population > x
    cur = conn.cursor()
    cur.execute("SELECT * FROM country WHERE Region = ? AND Population > ?", (region, population))
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
```

```python
select_asian_countries(conn, 'ASIA (EX. NEAR EAST)         ',200000000)
```

```python
def min_IMR_countries(conn, no_of_countries):# Top regions with the minimum IMR rates
    cur = conn.cursor()
    cur.execute("SELECT Region, AVG(`Infant mortality (per 1000 births)`) FROM country GROUP BY Region ORDER BY AVG(`Infant mortality (per 1000 births)`) ASC LIMIT ?", (no_of_countries,))
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
```

```python
min_IMR_countries(conn, 5)
```

```python
def top_agrarian_countries(conn, agri_percent, no_of_regions): #Regions with the most countries depending on agriculture
    cur = conn.cursor()
    cur.execute("SELECT Region, COUNT(Country) FROM country WHERE Agriculture > ? GROUP BY Region ORDER BY COUNT(Country) DESC LIMIT ?", (agri_percent, no_of_regions,))
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
```

```python
top_agrarian_countries(conn, 0.15, 10) 
```
